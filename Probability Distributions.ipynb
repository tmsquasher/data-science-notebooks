## Discrete Probability Distributions
### X ~ Binomial(n,p)
P(x) is the probability of x heads in n independent flips with probability of success p. The PMF below plots the probability of x heads in 20 independent flips, each with a probability of success of 0.25. The mean is 5 and the variance is 3.75.

$Pr(X = x) = {n \choose x} p^{x} (1-p)^{n-x}$

$E(X) = np$

$Var(X) = np(1-p)$

```python
def binomial(n, p, x):
    choose = math.factorial(n) / math.factorial(x) / math.factorial(n - x)
    return choose * pow(p, x) * pow((1 - p), (n - x))

n = 20
ys_b = [binomial(n, 0.25, x) for x in xrange(0, n)]
ys_r = [binomial(n, 0.5, x) for x in xrange(0, n)]
p1, = plt.plot(xrange(0, n), ys_b, 'bo-', alpha=0.5)
p2, = plt.plot(xrange(0, n), ys_r, 'ro-', alpha=0.5)
plt.legend([p1, p2], ['p=0.25, n=20', 'p=0.5, n=20'])
plt.xlabel('x')
plt.ylabel('P(X=x)')
```

### X ~ Geometric(p)
Pr(X = x) is the probability that the first success in a series of flips with probability p occurs on trial x. There are (x - 1) trials without success, followed by a single trial with success.

$Pr(X = x) = p^x (1 - p)^{x - 1}$

$E(X) = 1 / p$

```python
def geometric(p, x):
    return p * pow((1 - p), (x - 1))

n = 20
ys_b = [geometric(0.25, x) for x in xrange(1, n)]
ys_r = [geometric(0.5, x) for x in xrange(1, n)]
p1, = plt.plot(xrange(1, n), ys_b, 'bo-', alpha=0.5)
p2, = plt.plot(xrange(1, n), ys_r, 'ro-', alpha=0.5)
plt.legend([p1, p2], ['p=0.25', 'p=0.5'])
plt.xlabel('x')
plt.ylabel('P(X=x)')
```

### X ~ Poisson($\lambda$)

Pr(X = x) is the probability of x successes in a given time interval that occur at an average rate of $\lambda$.

$Pr(X = x) = \dfrac {e^{-\lambda} \lambda^x} {x!}$

$E(X) = \lambda$

```python
def poisson(_lambda, x):
    return exp(-_lambda) * pow(_lambda, x) / math.factorial(x)

n = 20
ys_b = [poisson(4, x) for x in xrange(0, n)]
ys_r = [poisson(10, x) for x in xrange(0, n)]
p1, = plt.plot(xrange(0, n), ys_b, 'bo-', alpha=0.5)
p2, = plt.plot(xrange(0, n), ys_r, 'ro-', alpha=0.5)
plt.legend([p1, p2], ['$\lambda$=4', '$\lambda$=10'])
plt.xlabel('x')
plt.ylabel('P(X=x)')
```
