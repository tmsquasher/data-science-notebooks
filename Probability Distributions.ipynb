# Probability Distributions
## Discrete PMFs
**Binomial:** $P(x; n, p) = {n \choose x} p^{x} (1-p)^{n-x}$, probability of x successes in n Bernoulli trials with success probability p

**Geometric:** $P(x; p) = p^x (1 - p)^{x - 1}$, probability that the first success in a series of Bernoulli trials occurs on the xth trial

**Poisson:** $P(x; \lambda) = \dfrac {e^{-\lambda} \lambda^x} {x!}$, probability of x successes in a given time interval with known rate $\lambda$

```python
# discrete distributions: binomial, geometric, poisson 
plt.figure(figsize(18, 4))
alpha = 0.75
x_range = 21

# binomial
for p in [0.25, 0.5, 0.75]:
    plt.plot(xrange(x_range), scipy.stats.binom(20, p).pmf(xrange(x_range)), 'o-', alpha=alpha, label='$n=20, p=' + str(p) + '$')
plt.title('Binomial PMF')
plt.legend()
plt.show()

# geometric
for p in [0.25, 0.5, 0.75]:
    plt.plot(xrange(x_range), scipy.stats.geom(p).pmf(xrange(x_range)), 'o-', alpha=alpha, label='$p=' + str(p) + '$')
plt.title('Geometric PMF')
plt.legend()
plt.show()

# poisson
for lam in [2, 5, 10, 20]:
    plt.plot(xrange(x_range), scipy.stats.poisson(lam).pmf(xrange(x_range)), 'o-', alpha=alpha, label='$\lambda=' + str(lam) + '$')
plt.title('Poisson PMF')
plt.legend()
plt.show()
```

```python
# continuous distributions:
x_range = linspace(0, 20, 100)
plt.figure(figsize(18, 4))

# uniform
for a,b in [(5, 10), (2, 16)]:
    plt.plot(x_range, scipy.stats.uniform(a, b).pdf(x_range), '-', linewidth=2, 
             alpha=alpha, label='$a=' + str(a) + ', b=' + str(b) + '$')
plt.title('Uniform PDF')
plt.legend()
plt.show()

# exponential
for lam in [0, 1, 5, 10]:
    plt.plot(x_range, scipy.stats.expon(lam).pdf(x_range), '-', linewidth=2, 
             alpha=alpha, label='$\lambda=' + str(lam) + '$')
plt.title('Exponential PDF')
plt.legend()
plt.show()

# gaussian
for mu,sigma in [(10, 1), (10, 2), (10, 5)]:
    plt.plot(x_range, scipy.stats.norm(mu, sigma).pdf(x_range), '-', linewidth=2, 
             alpha=alpha, label='$\mu=10, \sigma=' + str(sigma) + '$')
plt.title('Gaussian PDF')
plt.legend()
plt.show()

# gamma
for a,b in [(1, 0.5), (1, 1), (1, 5), (2, 10)]:
    plt.plot(x_range, scipy.stats.gamma(a, b).pdf(x_range), '-', linewidth=2, 
             alpha=alpha, label='$a=' + str(a) + ', b=' + str(b) + '$')
plt.title('Gamma PDF')
plt.legend()
plt.show()

# beta
for a,b in [(0.5, 0.5), (0.2, 0.8), (5, 10), (1, 20)]:
    plt.plot(linspace(0, 1, 100), scipy.stats.beta(a, b).pdf(linspace(0, 1, 100)), '-', linewidth=2, 
             alpha=alpha, label='$a=' + str(a) + ', b=' + str(b) + '$')
plt.title('Beta PDF')
plt.legend()
plt.show()
```
