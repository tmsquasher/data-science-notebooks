# Naive Bayes Classifier
## Make Random Dataset

```python
# random dataset with two features (x1 / x2) and two classes (black / white)
X, y = make_classification(n_samples=200, n_features=2, n_classes=2, 
                           n_informative=1, n_redundant=0, n_repeated=0, 
                           n_clusters_per_class=1, random_state=10)
df = pd.DataFrame(X, columns=['x1', 'x2'])
df['black'] = y
x1_black = df.x1[df.black == 1]
x2_black = df.x2[df.black == 1]
x1_white = df.x1[df.black == 0]
x2_white = df.x2[df.black == 0]
plt.figure(figsize(12.5, 5))
plt.scatter(x1_black, x2_black, c='black', label='$black=1$')
plt.scatter(x1_white, x2_white, c='white', label='$black=1$')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Original Data')
plt.show()
```

## Gaussian Naive Bayes from Scratch
We want to create a classifier that predicts whether a dot is going to be black (y=1) or white (y=0), based on features $x_1$ and $x_2$.

#### 1. Start with Bayes' Theorem 
$p(y | x_1, x_2) = \frac{p(y)p(x_1, x_2 | y)}{p(x_1, x_2)}$ or, $posterior = \frac{prior * likelihood}{evidence}$. In our case, the posterior is the probability that a dot is black, given its features x1 and x2; the prior is our prior belief for what color the dot would be (regardless of the x1 and x2 features); the likelihood is the probability of the features, given the color; and the evidence is the total unconditional probability of the features we got. For our purposes, the evidence is constant and can be ignored.

#### 2. Assume Feature Independence
If we assume that the features are COMPLETELY independent of each other, then the likelihood term can be broken into $p(x_1, x_2 | y) = p(x_1 | y)p(x_2 | y)$. This strong assumption is why this classifier is called 'naive.' Using this assumption and dropping the evidence denominator, our estimate of a dot's color becomes:

$p(y | x_1, x_2) = p(y)*p(x_1|y)p(x_2|y)$

#### 3. Model the Prior and Likelihood
The prior $p(y)$ is estimated as the fraction of dots in our training data that are black -- this makes sense because this is all the information we have, and if there are a lot of blacks, then we have reason to believe incoming dots will be, too.

The likelihood pieces $p(x_1|y)p(x_2|y)$ are modeled as Gaussians with $\mu$ and $\sigma$ as the means and standard deviations of the features for the observations with the corresponding values of y. For example, the for the likelihood piece $p(x_1|y=black)$, we generate a Gaussian with $\mu = mean(x_1 \text{ where } y=black)$, and $\sigma = std(x_1 \text{ where } y=black)$ -- then for each actual data point, we see how probable its $x_1$ would be in this model. 


```python
# priors
prior_black = float(df.black.sum()) / len(df)
prior_white = 1 - prior_black

# likelihoods
plt.figure(figsize(12.5, 5))
plt.scatter(x1_black, x2_black, c='black')
plt.scatter(x1_white, x2_white, c='white')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Feature Means for Each Color')
plt.xlim(-4, 3)
plt.ylim(-3, 4)
plt.vlines(x1_black.mean(), -3, 4, color='red', linestyles='solid', label='$\mu (x_1, y=black)$')
plt.vlines(x1_white.mean(), -3, 4, color='blue', linestyles='solid', label='$\mu (x_1, y=white)$')
plt.hlines(x2_black.mean(), -4, 3, color='red', linestyles='dashed', label='$\mu (x_2, y=black)$')
plt.hlines(x2_white.mean(), -4, 3, color='blue', linestyles='dashed', label='$\mu (x_2, y=white)$')
plt.legend()
plt.show()
```

#### 4. Predict Probabilities
Training the classifier is fast and is powerful with lots of features/dimensions -- the means and variances going into the each feature's Gaussian model don't change unless we add new data. For each training point, calculate the $p(y=black | x_1, x_2)$ by multiplying the corresponding prior (in our case, 0.5 since our training data is evenly split between black and white) by the corresponding likelihood (multiply the Gaussian probability for each feature). 

Do the same for $p(y=white | x_1, x_2)$, and normalize these two probabilities to be [0, 1]. Classify the dot by whichever probability is bigger!

```python
# gaussian probability
def gaussianLikelihood(x, mu_class, sigma_class):
    return 1 / sqrt(2 * np.pi * sigma_class**2) * np.exp(-(x - mu_class)**2 / (2 * sigma_class**2))

# posterior = prior * likelihood
df['p_black'] = df.apply(lambda row: 
         prior_black * 
         gaussianLikelihood(row['x1'], x1_black.mean(), x1_black.std()) * 
         gaussianLikelihood(row['x2'], x2_black.mean(), x2_black.std()), axis=1)
df['p_white'] = df.apply(lambda row: 
         prior_white * 
         gaussianLikelihood(row['x1'], x1_white.mean(), x1_white.std()) * 
         gaussianLikelihood(row['x2'], x2_white.mean(), x2_white.std()), axis=1)

# normalize probabilities
df['p_black'] = df.p_black / (df.p_black + df.p_white)
df['p_white'] = 1 - df.p_black
df['prediction'] = df.apply(lambda row: 1 if row['p_black'] > row['p_white'] else 0, axis=1)
```
