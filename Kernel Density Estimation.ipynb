# Kernel Density Estimation
** A kernel function is any symmetric function that integrates to 1. Kernel density estimation is used to estimate the PDFs behind the distributions (histograms) of sampled data. It typically involves smoothing the histogram by summing kernels at each point. Here, we'll look at the case of Gaussian kernels. **

```python
import numpy as np
import scipy as sp
from sklearn.neighbors.kde import KernelDensity
%pylab inline
```

```python
# test data: a few samples from a poisson random variable
data = [sp.random.poisson(5) for x in xrange(10)]
```

## Gaussian Kernel Density Estimation
** Gaussian PDF: **
$P(x) = \frac{1}{\sqrt{ 2 \pi \sigma^2 }} e^{ - \frac{ (x - \mu)^2 } {2 \sigma^2} }, \text{ normal when } \mu=0 \text{ and } \sigma=1$

** Gaussian Kernel: **
$K(x; h) = exp(-\frac{x^2}{2h^2}), \text{ where } h \text{ is the bandwidth; effectively }\sigma \text{ from the PDF} $

** Kernel Density Estimator: **
$f(x) = \frac{1}{n}\sum_{i=1}^n{K_h(x - x_i)} = \frac{1}{nh}\sum_{i=1}^n{\frac{K_h(x - x_i)}{h}}, \text{ where the } x_i \text{'s are all the points from the original data, and K is the kernel function} $

```python
### Gaussian PDF
def p_gaussian(x, mu, sigma):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))

def pdf_gaussian(xs, mu, sigma):
    return [p_gaussian(x, mu, sigma) for x in xs]

plt.figure(figsize=(18, 3))
ax1 = plt.subplot2grid((1, 2), (0, 0))
ax1.plot(linspace(-5, 5), pdf_gaussian(linspace(-5, 5), 0, 1), label='$\sigma=1$')
ax1.plot(linspace(-5, 5), pdf_gaussian(linspace(-5, 5), 0, 2), label='$\sigma=2$')
ax1.set_title('Gaussian PDF', fontsize=20)
ax1.legend()


### Gaussian Kernel
def kernel_gaussian(x, h):
    return np.exp(-x**2 / (2*h**2))

ax2 = plt.subplot2grid((1, 2), (0, 1))
ax2.plot(linspace(-5, 5), kernel_gaussian(linspace(-5, 5), 1), label='$h=1$')
ax2.plot(linspace(-5, 5), kernel_gaussian(linspace(-5, 5), 2), label='$h=2$')
ax2.set_title('Gaussian Kernel', fontsize=20)
ax2.legend()
plt.show()


### Kernel Density Estimator
plt.figure(figsize=(18, 4))

# plot individual kernels
bandwidth = 1.0
sigma = 1.0
ax1 = plt.subplot2grid((1, 1), (0, 0))
ax1.hist(data, normed=True, alpha=0.5, label='Original histogram')
first = True
for x in data:
    kernel_xs = linspace(x - 5*sigma, x + 5*sigma)
    kernel_ys = np.array(pdf_gaussian(kernel_xs, x, sigma)) * 0.4 # scaling just so it looks better
    if first:
        ax1.plot(kernel_xs, kernel_ys, 'r--', linewidth=1, label='Kernels')
        first = False
    else:
        ax1.plot(kernel_xs, kernel_ys, 'r--', linewidth=1)        

# plot kernel density estimator (sum of individual kernels)
min_data = min(data) - 4
max_data = max(data) + 4
ys = []
xs = linspace(min_data, max_data, 100)
for x in xs:
    kde = sum([kernel_gaussian(x - xi, bandwidth) for xi in data]) / (len(data) * bandwidth)
    ys.append(kde)
ax1.plot(xs, ys, linewidth=3, color='blue', label='KDE, $h=1$')
ax1.set_xlim(min_data, max_data)
ax1.set_ylim(0, 0.4)
ax1.set_title('Kernel Density Estimator, $h=1$', fontsize=20)
ax1.legend()
plt.show()


### Second KDE - basically copy/pasted code below here
# plotting one more kde for h = 2
plt.figure(figsize=(18, 4))
bandwidth = 2.0
sigma = 2.0
ax1 = plt.subplot2grid((1, 1), (0, 0))
ax1.hist(data, normed=True, alpha=0.5, label='Original histogram')
first = True
for x in data:
    kernel_xs = linspace(x - 5*sigma, x + 5*sigma)
    kernel_ys = np.array(pdf_gaussian(kernel_xs, x, sigma)) * 0.4 # scaling just so it looks better
    if first:
        ax1.plot(kernel_xs, kernel_ys, 'r--', linewidth=1, label='Kernels')
        first = False
    else:
        ax1.plot(kernel_xs, kernel_ys, 'r--', linewidth=1)  
ys = []
for x in xs:
    kde = sum([kernel_gaussian(x - xi, bandwidth) for xi in data]) / (len(data) * bandwidth)
    ys.append(kde)
ax1.plot(xs, ys, linewidth=3, label='KDE, $h=2$')
ax1.set_xlim(min_data, max_data)
ax1.set_ylim(0, 0.4)
ax1.set_title('Kernel Density Estimator, $h=2$', fontsize=20)
ax1.legend()
plt.show()
```

** Key points: **

** - The Gaussian PDF and the Gaussian Kernel are identical, save for some scaling **

** - The red kernels are summed/scaled to produce the KDE ** 

** - higher $\sigma$/h/binning factor creates more smoothing -- more bias, less variance. **
