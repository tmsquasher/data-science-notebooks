## Central Limit Theorem
** Pick a sample size $n$ and ANY population distribution. The frequencies of sample means (i.e., the sampling distribution) will be distributed normally around the population mean with standard deviation $SD=\sigma / \sqrt{n}$, where $\sigma$ is the standard deviation of the original population distribution. In general, $n=30$ leads to a close enough approximation to the normal distribution for us to make inferences (the t-distribution, which effectively IS our estimate of the sampling distribution, very closely approximates the normal distribution when it has 30+ degrees of freedom). **

```python
plt.figure(figsize=(18, 4))
alpha = 0.5
beta = 0.5

# original distribution is beta with wacky bimodal shape
ax1 = plt.subplot2grid((1, 2), (0, 0))
ax1.hist(np.random.beta(alpha, beta, 5000), normed=True, bins=20)
ax1.set_title('Samples from Population Distribution')
ax1.vlines(0.5, 0, 3, colors='r', linestyles='dashed', label='$\mu=0.5$')
ax1.set_ylim(0, 3)
ax1.legend()

# the sampling distribution comes from histogramming the means of lots of samples of the same size
n = 30
means = [np.random.beta(alpha, beta, n).mean() for x in xrange(10000)]
ax2 = plt.subplot2grid((1, 2), (0, 1))
ax2.hist(means, bins=50, normed=True)
ax2.set_title('Sampling Distribution of the Mean, $n=30$')
ax2.vlines(0.5, 0, 700, colors='r', linestyles='dashed', label='$\mu=0.5$')
ax2.set_ylim(0, 10.0)

# plot standard error of the mean
beta_variance = (alpha * beta) / ((alpha + beta)**2 * (alpha + beta + 1))
stderr = math.sqrt(beta_variance / n)
ax2.hlines(3.7, 0.5, 0.5 + stderr, colors='green', label=('$\sigma/n=' + str(round(stderr, 3)) + '$'))

# plot actual normal curve
def gauss2d(x): return 1/(np.sqrt(2*np.pi)*stderr) * np.exp(-((x-0.5)**2)/(2*stderr**2))
ax2.plot(linspace(0.2, 0.8), gauss2d(linspace(0.2, 0.8)), linewidth=3, alpha=0.8)
ax2.legend()
plt.show()
```

## Sampling Distributions for Various Sample Sizes ($n$)

```python
ns = [1, 2, 10, 30, 100, 1000, 10000]
def plotIt(n, k):
    means = [np.random.beta(alpha, beta, n).mean() for x in xrange(10000)]
    plt.figure(figsize=(10, 3))
    ax2 = plt.subplot2grid((1, 1), (0, 0))
    ax2.hist(means, bins=50, normed=True)
    
    beta_variance = (alpha * beta) / ((alpha + beta)**2 * (alpha + beta + 1))
    stderr = math.sqrt(beta_variance / n)
    
    ax2.set_title('$n=' + str(n) + '$, ' + ('$\sigma/n=' + str(round(stderr, 3)) + '$'))
    ax2.set_xlim(0, 1.0)
    plt.show()

for k in xrange(len(ns)):
    plotIt(ns[k], k)
```
