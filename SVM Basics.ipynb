## SVM with Linear, RBF, and Polynomial Kernels

```python
def plotWithKernel(X, y, kern, title, c=1.0, g=0.0, size=(8, 4)):
    # create classifier
    clf = sklearn.svm.SVC(kernel=kern, C=c, gamma=g)
    clf.fit(X, y)
    
    plt.figure(figsize=size)
    ax = plt.axes()
    ax.set_xlim(-3, 2)
    ax.set_ylim(-3, 3)
    ax.set_title(title)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    
    # plot mesh
    step_size = 0.01
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, step_size), np.arange(y_min, y_max, step_size))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=plt.cm.Paired)
    
    # plot original scatter
    ax.scatter([x[0] for x in clf.support_vectors_], [x[1] for x in clf.support_vectors_], s=80, c='red', alpha=1, linewidths=0) 
    ax.scatter([x[0] for x in X], [x[1] for x in X], c=y, cmap='gray')
    plt.show()

plotWithKernel(X, y, 'linear', 'Linear Kernel Hyperplane and Support Vectors')
plotWithKernel(X, y, 'rbf', 'RBF Kernel Hyperplane and Support Vectors')
plotWithKernel(X, y, 'poly', '3-Degree Polynomial Kernel Hyperplane and Support Vectors')
```

## RBF Kernel with Various Gamma Hyperparameters

```python
plotWithKernel(X, y, 'rbf', 'RBF Kernel, Gamma=0.0', g=0.0)
plotWithKernel(X, y, 'rbf', 'RBF Kernel, Gamma=0.1', g=0.1)
plotWithKernel(X, y, 'rbf', 'RBF Kernel, Gamma=0.1', g=10)
```

## RBF Kernel with Various C Hyperparameters

```python
plotWithKernel(X, y, 'rbf', 'RBF Kernel, C=1.0', c=1.0)
plotWithKernel(X, y, 'rbf', 'RBF Kernel, C=10.0', c=10.0)
plotWithKernel(X, y, 'rbf', 'RBF Kernel, C=100.0', c=100.0)
```
