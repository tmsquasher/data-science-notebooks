# Model Fitting
## Maximum Likelihood Estimation
In a nutshell, Maximum Likelihood Estimation (MLE) finds the parameters of a model that most closly fit your observed data. This notebook is going to be the anti-theory approach to understanding MLE.

### Single Parameter Estimation: Simple Coin-Flip Example
Suppose you're given a strange coin. It might be fair, it might be unfair, we don't know. All we know is that it will come up 'heads' with an unknown probability $\theta$. We can use Maximum Likelihood Estimation to come up with our best estimate for $\theta$ by flipping the coin for $n$ trials.

For maximum likelihood estimation of $\theta$, we assume that our datapoints $X_1, X_2, ..., X_n$ are drawn randomly (i.i.d., or "independently and identically distributed"). Because they are drawn independently, the probability of all our datapoints given $\theta$ is just the product of the probabilities of each one, given $\theta$. (This is one reason why random sampling is so important -- we need it for MLE's independence assumption to hold.)

$Likelihood(\theta) = P(X_1, X_2, ... X_n|\theta) = \prod_{i=1}^{n}{P(X_i|\theta)}$

So, our best estimate of $\theta$ is just the value that MAXIMIZES the LIKELIHOOD function above. While we can sometimes calculate this analytically, for our purposes, we'll see that we can naively brute force a lot of values for $\theta$ and come up with a pretty good estimation that way, too.

```python
# set an unknown, actual P(heads) for our coin
# we will try to estimate this based on our generated data
THETA_ACTUAL = 0.66

# generate data: 1=heads, 0=tails, 100 trials
data = np.random.binomial(1, THETA_ACTUAL, 100) 

# likelihood calculation, given a theta and some data
def likelihood(theta, data):
    xs = []
    for x in data:
        if x == 1: 
            # datapoint is heads --> what's the probability of that, given theta?
            p_x_given_theta = theta
        else:
            # datapoint is tails --> what's the probability of that, given theta?
            p_x_given_theta = 1 - theta
        xs.append(p_x_given_theta)
    return np.array(xs).prod()

# brute-force: get theta that maximizes likelihood
thetas_to_try = linspace(0, 1, 100)
likelihoods = []
maximum = -1
best_theta = 0
for theta in thetas_to_try:
    L = likelihood(theta, data)
    likelihoods.append(L)
    if L > maximum:
        maximum = L
        best_theta = theta

# plot our results
plt.figure(figsize=(18, 5))
plt.bar(thetas_to_try, likelihoods, width = 0.005, alpha=0.75, linewidth=0)
plt.title('Maximum Likelihood: ' + str(max(likelihoods)) + ' at ' + str(best_theta))
plt.xlabel('A guess at theta; i.e., P(heads)')
plt.ylabel('Likelihood')
plt.vlines(THETA_ACTUAL, 0, max(likelihoods), colors='r', linewidth=2, label='THETA_ACTUAL')
plt.ylim(0, max(likelihoods))
plt.legend()
plt.show()

print 'Total heads in data:', sum(data)
print 'Analytical MLE:', float(sum(data)) / len(data)
print 'Brute force MLE:', best_theta
print '------------------'
print 'THETA_ACTUAL:', THETA_ACTUAL
```

** A few takeaways: **

1. The likelihood values we calcualted are TINY. Like, 1e-27 tiny. For this and other reasons, we often use 'Log-Likelihood' rather than pure Likelihood in our calculations. (i.e., instead of getting values like 1e-27, we'd get values like -27)

2. The analytical solution to the problem above is incredibly trivial: we just count the number of heads in our dataset, and divide that by the total flips. That's the best estimation we have for P(heads)! Our brute-force computational approach was completely unnecessary, but hopefully illustrated the method clearly.

### Two Parameter Estimation: Fitting a Gaussian Model

Yes, there is an easy analytical way to fit a set of data to the Gaussian distribution (set the derivative of the Log-Likelihood function to 0, then solve directly for $\mu$ and $\sigma$). But let's try to brute-force it anyways for visualization purposes. (And yes, there are better ways to search the parameter space than simple brute-force; bear with me!)

We're going to use Log-Likelihood this time to make the numbers a little more tractible. So, instead of taking the product of all the likelihoods for each data point, we take the sum of all the log-likelihoods for each data point. 

```python
from scipy.stats import norm

# generate data: 100 i.i.d. samples
MU_ACTUAL = 15
SIGMA_ACTUAL = 3
data = np.random.normal(MU_ACTUAL, SIGMA_ACTUAL, 100)

def gaussianPDF(mu, sigma, x):
    return 1./(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) )

def logLikelihoodGaussian(mu, sigma, data):
    pxs = []
    for x in data:
        p_x = log(gaussianPDF(mu, sigma, x))
        pxs.append(p_x)
    return np.array(pxs).sum()

# brute force: try a lot of mu's and sigma's
test_mus = linspace(5, 25, 50)
test_sigmas = linspace(1, 5, 50)
likelihoods = []
mus = []
sigmas = []
maximum_likelihood = -np.inf
best_mu = -1
best_sigma = -1
for mu in test_mus:
    for sigma in test_sigmas:
        L = logLikelihoodGaussian(mu, sigma, data)
        likelihoods.append(L)
        mus.append(mu)
        sigmas.append(sigma)
        if L > maximum_likelihood:
            maximum_likelihood = L
            best_mu = mu
            best_sigma = sigma

# plot results
plt.figure(figsize=(18, 6))
plt.scatter(mus, sigmas, c=likelihoods, s=500, linewidths=0, cmap=plt.cm.coolwarm)
plt.scatter(best_mu, best_sigma, c='yellow', s=100, linewidth=0, label='mu and sigma combo with highest log-likelihood')
plt.xlim(5, 25)
plt.ylim(1, 5)
plt.title('Log-Likelihood values (color) for various $\mu$ and $\sigma$')
plt.xlabel('mu')
plt.ylabel('sigma')
plt.legend()
plt.show()

plt.figure(figsize=(18, 6))
plt.hist(data, bins=20, normed=True, alpha=0.75)
plt.plot(linspace(5, 25), plt.mlab.normpdf(linspace(5, 25), best_mu, best_sigma), linewidth=4, c='r', label='Fitted')
plt.plot(linspace(5, 25), plt.mlab.normpdf(linspace(5, 25), MU_ACTUAL, SIGMA_ACTUAL), linewidth=2, c='y', label='Actual (Unknown)')
plt.title('Original data distribution/histogram with fitted Gaussian overlay')
plt.show()

print 'Fitted mu:', best_mu
print 'Fitted sigma:', best_sigma
print '-----------------'
print 'Actual mu:', MU_ACTUAL
print 'Actual sigma:', SIGMA_ACTUAL
```

** Takeaways **

In general, the Likelihood is just the probability you'd get the data you did, given your model and parameters. All you have to do is try a bunch of different parameter values for your model, multiply the values (or sum the log-values) you get from your model's PDF (I suppose your 'model' is just a PDF, after all). Your "fitted" model is just the set of parameters that maximize this Likelihood calculation.
